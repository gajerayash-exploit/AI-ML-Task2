```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("titanic.csv")
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.describe(include='object')
```

```python
df.hist(figsize=(12, 10), bins=20)
plt.tight_layout()
plt.show()
```

```python
num_cols = df.select_dtypes(include=np.number).columns

plt.figure(figsize=(12, 6))
sns.boxplot(data=df[num_cols])
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.pairplot(df[['Survived', 'Pclass', 'Age', 'Fare']])
plt.show()
```

```python
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.show()
```

```python
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title("Survival by Gender")
plt.show()
```

```python
sns.countplot(x='Survived', hue='Pclass', data=df)
plt.title("Survival by Passenger Class")
plt.show()
```

# Conclusion

Exploratory Data Analysis (EDA) was performed on the Titanic dataset using statistical summaries and visualizations.

Key insights:
- Most passengers did not survive.
- Female passengers had higher survival chances.
- Passenger class affected survival probability.
- Fare and Age showed some outliers.
- Correlation analysis helped identify relationships between features.

EDA helped in understanding the dataset structure, trends, distributions, and important feature relationships before applying machine learning models.
